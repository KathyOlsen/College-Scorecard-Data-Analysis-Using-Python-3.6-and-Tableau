```python
def relaffil(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
df['relaffil'] = df['RELAFFIL'].apply(relaffil)
df.head()
```

```python
df.to_csv('df1_2015_16_w_relaffil.csv',index=False)
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='OMAWDP8_FTFT ~ WOMENONLY + relaffil + ADM_RATE + SAT_AVG + AVGFACSAL + PCTFLOAN - 1',data=df).fit()
result.summary()
```
